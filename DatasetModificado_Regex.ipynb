```python
# importando as Bibliotecas
import pandas as pd
import re
```

```python
# Importando o Dataset
df=pd.read_excel('/content/Dataset_Regex.xlsx')
df
```

```python
# Função para extrair servidor de email
def extrair_servidor_email(email):
    match = re.search(r'@(\w+)\.', email)
    if match:
        return match.group(1)
    else:
        return None

# Função para extrair dia, mês e ano
def extrair_data(data):
    match = re.match(r'(\d{2})/(\d{2})/(\d{4})', data)
    if match:
        return match.group(1), match.group(2), match.group(3)
    else:
        return None, None, None

# Função para extrair código de área
def extrair_area(telefone):
    match = re.search(r'\((\d{2})\)', telefone)
    if match:
        return match.group(1)
    else:
        return None

# Função para extrair nome do site
def extrair_site(url):
    match = re.search(r'\.(\w+)\.', url)
    if match:
        return match.group(1)
    else:
        return None
```

```python
# Aplicar as funções aos respectivos campos do DataFrame
df['servidor_email'] = df['Email'].apply(extrair_servidor_email)
df['dia'], df['mes'], df['ano'] = zip(*df['Data'].apply(extrair_data))
df['area'] = df['Telefone'].apply(extrair_area)
df['site'] = df['URL'].apply(extrair_site)

# Exibir o DataFrame resultante
df_modificado=pd.DataFrame(df)
df_modificado.head(10)
```
